```python
import datetime
import pandas as pd
pd.set_option('expand_frame_repr', False)

df = pd.read_csv('BTC-USD-1d.csv')
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df.dtypes
```

```python
df[df['Date'] == '2013-04-02']
df.sample(5) # random 5 records
```

```python
df = df[['Date', 'Close']]
```

```python
# define start and end date
startdate = '2013-12-04'
enddate = '2019-12-31'
df = df[(df['Date'] >= startdate) & (df['Date'] <= enddate)]
```

```python
# calculate in amount

df['everytime_in'] = 100 # every week put in 100
df['accumulate_in'] = df['everytime_in'].cumsum() # cumulative_sum

c_rate = 0.002 # 手续费

df['everytime_buy_bt'] = df['everytime_in'] / df['Close'] * (1 - c_rate)
df['acc_buy_bt'] = df['everytime_buy_bt'].cumsum() 
```

```python
df
```

```python
# calculate market price
df['average_cost'] = df['accumulate_in'] / df['acc_buy_bt']
df['acc_value'] = df['acc_buy_bt'] * df['Close']
df
```
