```python
import seaborn as sns
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def sinplot(flip=1):
    x = np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x, np.sin(x+1*0.5) * (7-i)*flip)
    
sinplot()
```

```python
sns.set()
sinplot()
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set(rc={"figure.figsize":(6,6)})
```

```python
current_palette = sns.color_palette()
sns.palplot(current_palette)
```

```python
# 6 default color themes: deep, muted, pastel, bright, dark, colorblind
sns.palplot(sns.color_palette("hls", 10))
```

```python
data = np.random.normal(size=(20,8)) + np.arange(8) /2
sns.boxplot(data=data, palette=sns.color_palette("hls", 10))
```

```python
# l = lightness, s = saturation
sns.palplot(sns.hls_palette(8,l=.5,s=.7))
```

```python
# colors use for compare
sns.palplot(sns.color_palette("Paired",10))
```
